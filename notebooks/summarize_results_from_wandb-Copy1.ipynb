## Summarize results from CIFAR-100

* There are 10 tasks
* Each task has to learn 10 classes
* Classes from one task do not appear in other tasks
* The model knows from which taks a sample belongs (This scenario is knwons as Task Incremental Learning)

```python
api = wandb.Api()
runs = api.runs("cf-lora-cifar100")
```

```python
times = {
    "lora": [],
    "sena": []
}

for i in range(20):
    print(runs[i].name, runs[i].history().iloc[-1]._runtime)
    if runs[i].name.__contains__("-lora-"):
        times["lora"].append(runs[i].history().iloc[-1]._runtime)
    else:
        times["sena"].append(runs[i].history().iloc[-1]._runtime)
```

```python
results = {}

for key, values in times.items():
    trimmed = values.copy()
    trimmed.remove(max(trimmed))
    trimmed.remove(min(trimmed))
    mean = statistics.mean(trimmed)
    stdev = statistics.stdev(trimmed)
    results[key] = {'mean': mean, 'stdev': stdev}

results
```

```python
val_acc_by_experiment_cifar100 = {'lora': [], 'sena': []}
last_run_cifar100 = {'lora': None, 'sena': None}

api = wandb.Api()
runs = api.runs("cf-lora-cifar100")

for run in runs:
    run_name = run.name
    best_values = []
    init_values = []
    num_epochs = []
    
    best_val=-1.
    init_val=-1.
    number_of_epochs=0
    
    hist = run.history()
    best_val_acc=hist['best_val_acc'].values
    for v in best_val_acc:
        if v < best_val:
            best_values.append(best_val)
            init_values.append(init_val)
            num_epochs.append(number_of_epochs)
            
            best_val=-1.
            init_val=-1
            number_of_epochs=0
        else:
            best_val = v
        
        number_of_epochs += 1
        if number_of_epochs <= value_considered_as_fewshot and v > init_val:
            init_val = v

    best_values.append(best_val_acc[-1])
    init_values.append(init_val)
    num_epochs.append(number_of_epochs)
    
    key_exp = 'lora' if '-lora-' in run_name else 'sena'
    if key_exp == 'lora': # Has to continue and not process, because the the results of lora will be taken from runs in 'cf-lora-cifar-100-loravgg19'
        continue
    print(run_name)
    if '-9' in run_name:
        last_run_cifar100[key_exp] = run
        
    val_acc_by_experiment_cifar100[key_exp].append({
        "best_values": best_values,
        "init_values": init_values,
        "num_epochs": num_epochs,
    })


runs = api.runs("cf-lora-cifar-100-loravgg19")

for run in runs:
    run_name = run.name
    
    best_values = []
    init_values = []
    num_epochs = []
    
    best_val=-1.
    init_val=-1.
    number_of_epochs=0
    
    hist = run.history()
    best_val_acc=hist['best_val_acc'].values
    for v in best_val_acc:
        if v < best_val:
            best_values.append(best_val)
            init_values.append(init_val)
            num_epochs.append(number_of_epochs)
            
            best_val=-1.
            init_val=-1
            number_of_epochs=0
        else:
            best_val = v

        number_of_epochs += 1
        if number_of_epochs <= value_considered_as_fewshot and v > init_val:
            init_val = v

    best_values.append(best_val_acc[-1])
    init_values.append(init_val)
    num_epochs.append(number_of_epochs)
    
    key_exp = 'lora'
    print(run_name)
    if '-9' in run_name:
        last_run_cifar100[key_exp] = run
        
    val_acc_by_experiment_cifar100[key_exp].append({
        "best_values": best_values,
        "init_values": init_values,
        "num_epochs": num_epochs,
    })
```

## Comparsion of Params Grouth with VGG-19 Network Archtecture

```python
import matplotlib.pyplot as plt

"""
Lora 
Per task
 * in classifier: 1.940.680
 * in convolutional layers: 850.419
 ** Total: 2.791.099

Sena
* in classifier: 120.365.256
 * in convolutional layers: 5.735.800  -- 13.110.400
 ** Total: 126.101.056

"""

# Dados de exemplo
num_tarefas = [i for i in range(1,6)]  # Número de tarefas aprendidas
parametros_A_treinamento = [2791]  # Número de parâmetros de treinamento para o Modelo A
parametros_A_treinamento.extend([2791 for _ in range(len(num_tarefas)-1)])
parametros_B_treinamento = [126101]  # Número de parâmetros de treinamento para o Modelo B
parametros_B_treinamento.extend([126101 for _ in range(len(num_tarefas)-1)])



parametros_A_congelados = [0]  # Número de parâmetros congelados para o Modelo A
for i in range(len(num_tarefas)-1):
    parametros_A_congelados.append(parametros_A_congelados[i]+parametros_A_treinamento[i])
                                   
parametros_B_congelados = [0]  # Número de parâmetros congelados para o Modelo B
for i in range(len(num_tarefas)-1):
    parametros_B_congelados.append(parametros_B_congelados[i]+parametros_B_treinamento[i])




# Altura das barras inferiores (parametros congelados)
bottom_A = parametros_A_congelados
bottom_B = parametros_B_congelados

# Altura das barras superiores (parametros de treinamento)
top_A = [param_a + param_t for param_a, param_t in zip(parametros_A_congelados, parametros_A_treinamento)]
top_B = [param_b + param_t for param_b, param_t in zip(parametros_B_congelados, parametros_B_treinamento)]

# Largura das barras
width = 0.35

# Criando o gráfico
fig, ax = plt.subplots()

# Adicionando as barras
ax.bar([(n - 0.02) for n in num_tarefas], parametros_A_congelados, width, label='Params of Lora Method frozen from old tasks', hatch='||', color='darkgray')
ax.bar([(n - 0.02) for n in num_tarefas], parametros_A_treinamento, width, bottom=bottom_A, label='Params of Lora Method in training', hatch='||', color='r')

ax.bar([(n + width+0.02) for n in num_tarefas], parametros_B_congelados, width, label='Params of Sena Method frozen from old tasks', hatch='-', color='lightgray')
ax.bar([(n + width+0.02) for n in num_tarefas], parametros_B_treinamento, width, bottom=bottom_B, label='Params of Sena Method in training', hatch='-', color='y')

# Adicionando rótulos e título
ax.set_xlabel('Tasks Learned')
ax.set_ylabel('Number of Parameters of model (in thousands)')
ax.set_title('Growth Comparison: Lora vs. Sena Models in VGG19')
ax.set_xticks([n + width / 2 for n in num_tarefas])
ax.set_xticklabels(num_tarefas)
ax.legend()

# Exibindo o gráfico
plt.savefig('growth-comparsion-vgg19-all.png', dpi=300, transparent=True)
plt.show()
```
